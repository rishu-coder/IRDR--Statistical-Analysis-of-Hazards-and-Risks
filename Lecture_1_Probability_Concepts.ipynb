```python
import numpy as np

# Given probabilities
P_E = 2e-3  # Probability of a significant earthquake occurring
P_noE = 1 - P_E  # Probability of no significant earthquake
P_A_given_E = 1 - 0.15  # Probability of an alarm given an earthquake (true positive rate)
P_noA_given_noE = 1 - 0.15  # Probability of no alarm given no earthquake (true negative rate)
P_A_given_noE = 0.15  # Probability of an alarm given no earthquake (false positive rate)
P_noA_given_E = 0.15  # Probability of no alarm given an earthquake (false negative rate)

# 1. Probability that no significant earthquake takes place
print(f"P(noE) = {P_noE:.5f}")

# 2. Probability that a real-time alarm is issued given that an earthquake takes place
print(f"P(A|E) = {P_A_given_E:.5f}")

# 3. Probability that our system issues an alarm and that an earthquake takes place
P_A_and_E = P_E * P_A_given_E
print(f"P(A and E) = {P_A_and_E:.7f}")

# 4. Probability that our system issues an alarm and that no earthquake takes place
P_A_and_noE = P_noE * P_A_given_noE
print(f"P(A and noE) = {P_A_and_noE:.5f}")

# 5. Probability that our system issues an alarm (Total Probability Theorem)
P_A = P_A_and_E + P_A_and_noE
print(f"P(A) = {P_A:.7f}")

# 6. Probability that a significant earthquake takes place given that an alarm is issued (Bayes' Theorem)
P_E_given_A = P_A_and_E / P_A
print(f"P(E|A) = {P_E_given_A:.5f}")
```

```python
# prompt: Identification of Flood Prone Area An approximate way for identifying the flood-prone (FP) areas is to use a method called the topographic wetness index (TWI). This method identifies the FP areas as those areas with (TWI>t) where t is a given threshold. A more accurate way of calculating the flood prone areas consists of calculating the inundation profile. In this method, the inundated zones (IN) are identified as those with the flood profile h > 0. According to the TWI method, in a given area, there is 5% probability that the zone is flood prone. Moreover, the probability that a zone is not inundated (IN) given that it is indicated as flood prone FP is equal to 5% and the probability that a zone is not flood prone (FP) given that it is indicated as inundated IN is equal to 2%.
# • What is the probability that a given area is not flood prone?
# • What is the probability that a given area is inundated given that it is indicated as flood prone?
# • What is the probability that a given area is indicated both as flood prone and inundated?
# • What is the probability that a given area is not inundated but (and) it is flood prone?
# • Suppose that the probability that a zone is not inundated given that it is not flood prone is equal to 0.90. What is the probability that a given zone is inundated? (hint: consider using the total probability theorem)

# Given probabilities
P_FP = 0.05  # Probability that a zone is flood prone
P_not_IN_given_FP = 0.05  # Probability that a zone is not inundated given it's flood prone
P_not_FP_given_IN = 0.02  # Probability that a zone is not flood prone given it's inundated
P_not_IN_given_not_FP = 0.90 # Probability that a zone is not inundated given it's not flood prone


# 1. Probability that a given area is not flood prone
P_not_FP = 1 - P_FP
print(f"1. Probability of not flood prone: {P_not_FP:.4f}")

# 2. Probability that a given area is inundated given that it is indicated as flood prone
P_IN_given_FP = 1 - P_not_IN_given_FP
print(f"2. Probability of inundated given flood prone: {P_IN_given_FP:.4f}")

# 3. Probability that a given area is indicated both as flood prone and inundated
P_FP_and_IN = P_FP * P_IN_given_FP
print(f"3. Probability of flood prone and inundated: {P_FP_and_IN:.4f}")

# 4. Probability that a given area is not inundated but is flood prone
P_not_IN_and_FP = P_FP * P_not_IN_given_FP
print(f"4. Probability of not inundated and flood prone: {P_not_IN_and_FP:.4f}")

# 5. Probability that a given zone is inundated (using total probability theorem)
# P(IN) = P(IN|FP) * P(FP) + P(IN|not FP) * P(not FP)
P_IN_given_not_FP = 1 - P_not_IN_given_not_FP
P_IN = (P_IN_given_FP * P_FP) + (P_IN_given_not_FP * P_not_FP)
print(f"5. Probability of inundated: {P_IN:.4f}")
```


3.	Structural Damage Identification
The probability that a structural member develops fatigue crack is 10-4. Suppose that a non-destructive inspection is performed on the structural member in order to check whether it has developed a fatigue crack. However, the inspection is subject to some error: That is, the probability that the inspection rejects (says that the member has developed fatigue crack) a member with no fatigue crack is 0.01 (given no fatigue crack), and the probability that the test accepts a member with a fatigue crack is 0.01 (given fatigue crack).

•	What is the probability that the member has fatigue crack?
Consider the following logical statements:
D ≡ inspection detects a member crack
Cr ≡ the crack exist


•	What is the probability that the inspection rejects a member (given) with fatigue crack?

•	What is the probability that the member has fatigue crack and it is also rejected by the inspection?

•	What is the probability that the inspection rejects the member?

•	What is the probability that the inspection accepts the member?

•	Imagine that the inspection has rejected the member. What is the probability that the member has developed a fatigue crack (that means with what probability you believe the test results?



```python
# prompt: Structural Damage Identification The probability that a structural member develops fatigue crack is 10-4. Suppose that a non-destructive inspection is performed on the structural member in order to check whether it has developed a fatigue crack. However, the inspection is subject to some error: That is, the probability that the inspection rejects (says that the member has developed fatigue crack) a member with no fatigue crack is 0.01 (given no fatigue crack), and the probability that the test accepts a member with a fatigue crack is 0.01 (given fatigue crack).
# • What is the probability that the member has fatigue crack? Consider the following logical statements: D ≡ inspection detects a member crack Cr ≡ the crack exist
# • What is the probability that the inspection rejects a member (given) with fatigue crack?
# • What is the probability that the member has fatigue crack and it is also rejected by the inspection?
# • What is the probability that the member has no fatigue crack and it is rejected by the inspection?
# • What is the probability that the inspection rejects the member?
# • What is the probability that the inspection accepts the member?
# • Imagine that the inspection has rejected the member. What is the probability that the member has developed a fatigue crack (that means with what probability you believe the test results?

# Given probabilities
P_Cr = 1e-4  # Probability of fatigue crack
P_noCr = 1 - P_Cr  # Probability of no fatigue crack
P_D_given_noCr = 0.01  # Probability of rejection given no crack (false positive)
P_noD_given_Cr = 0.01  # Probability of acceptance given crack (false negative)
P_D_given_Cr = 1 - P_noD_given_Cr # Probability of rejection given crack (true positive)


# 1. Probability that the member has fatigue crack
print(f"1. P(Cr) = {P_Cr:.6f}")

# 2. Probability that the inspection rejects a member given it has a fatigue crack
print(f"2. P(D|Cr) = {P_D_given_Cr:.2f}")

# 3. Probability that the member has fatigue crack and is rejected by the inspection
P_Cr_and_D = P_Cr * P_D_given_Cr
print(f"3. P(Cr and D) = {P_Cr_and_D:.8f}")

# 4. Probability that the member has no fatigue crack and is rejected by the inspection
P_noCr_and_D = P_noCr * P_D_given_noCr
print(f"4. P(noCr and D) = {P_noCr_and_D:.6f}")

# 5. Probability that the inspection rejects the member
P_D = P_Cr_and_D + P_noCr_and_D
print(f"5. P(D) = {P_D:.6f}")

# 6. Probability that the inspection accepts the member
P_noD = 1 - P_D
print(f"6. P(noD) = {P_noD:.6f}")

# 7. Probability that the member has a fatigue crack given the inspection rejected it
P_Cr_given_D = P_Cr_and_D / P_D
print(f"7. P(Cr|D) = {P_Cr_given_D:.6f}")
```

```python
# prompt: The weather forecasting problem Suppose that during the month of October on average there is 10% probability that it rains heavily within a time interval of 24 hours. If there is 80% probability that the roof will leak due to heavy rain,
# • What is the probability that it rains heavily around noon today? Consider the following logical statements: R ≡ it rains heavily within a time interval of 24 hours Lr ≡ the roof leak due to heavy rain
# • What is the probability that it rains heavily around noon today and the roof leaks?
# • What is the probability that it does not rain heavily around noon today however the roof leaks? Suppose that there is only 2% probability that the roof leaks when there has not been a heavy rain before.
# • What is the probability that that the roof will leak this afternoon?
# • Imagine that you have spent the whole day on the midterm exam and have not looked outside the window to see if it has rained. In the afternoon you notice that the roof of the classroom is leaking, what is the probability that it has rained earlier today?

# Given probabilities
P_R = 0.10  # Probability of heavy rain within 24 hours
P_Lr_given_R = 0.80  # Probability of roof leak given heavy rain
P_Lr_given_not_R = 0.02  # Probability of roof leak given no heavy rain

# 1. Probability that it rains heavily around noon today
print(f"1. Probability of heavy rain around noon: {P_R:.2f}")

# 2. Probability that it rains heavily around noon and the roof leaks
P_R_and_Lr = P_R * P_Lr_given_R
print(f"2. Probability of heavy rain and roof leak: {P_R_and_Lr:.3f}")

# 3. Probability that it does not rain heavily but the roof leaks
P_not_R = 1 - P_R
P_not_R_and_Lr = P_not_R * P_Lr_given_not_R
print(f"3. Probability of no heavy rain but roof leak: {P_not_R_and_Lr:.3f}")

# 4. Probability that the roof will leak this afternoon
P_Lr = P_R_and_Lr + P_not_R_and_Lr
print(f"4. Probability of roof leak: {P_Lr:.3f}")

# 5. Probability that it rained given the roof is leaking (Bayes' Theorem)
P_R_given_Lr = P_R_and_Lr / P_Lr
print(f"5. Probability of rain given roof leak: {P_R_given_Lr:.3f}")
```


5.	Structural Damage Detection

It is also known that the probability of having damage is equal to 1%.

•	The result of a test is that a defect exists. What is the probability that the damage exists?

```python
# prompt: Structural Damage Detection It is known that the error associated with a specific destructive test for identification of a specific damage (defect) is equal to 5%. This means that the probability that the test detects the damage given that the defect does not exist is equal to 5%, and the probability of not damage detection given that the defect exists is also equal to 5%. It is also known that the probability of having damage is equal to 1%.
# • What is the probability that no damage (defect) exists? Consider the following logical statements: V ≡ the test detects the defect (damage) D ≡ the damage exists
# • What is the probability that the test detects a defect and that the damage really exists?
# • What is the probability that the test does not detect a defect and that the defect does not exist?
# • What is the probability that damage is detected by the test?
# • The result of a test is that a defect exists. What is the probability that the damage exists?

# Given probabilities
P_D = 0.01  # Probability of damage
P_V_given_not_D = 0.05  # Probability of detecting damage given no damage (false positive)
P_not_V_given_D = 0.05  # Probability of not detecting damage given damage (false negative)

# 1. Probability that no damage exists
P_not_D = 1 - P_D
print(f"1. Probability of no damage: {P_not_D:.2f}")

# 2. Probability of detecting damage and damage exists
P_V_given_D = 1 - P_not_V_given_D  # Probability of detecting damage given damage (true positive)
P_V_and_D = P_D * P_V_given_D
print(f"2. Probability of detecting damage and damage exists: {P_V_and_D:.4f}")

# 3. Probability of not detecting damage and no damage
P_not_V_given_not_D = 1 - P_V_given_not_D # Probability of not detecting damage given no damage (true negative)
P_not_V_and_not_D = P_not_D * P_not_V_given_not_D
print(f"3. Probability of not detecting damage and no damage: {P_not_V_and_not_D:.4f}")

#4. Probability that damage is detected by the test
P_V = P_V_and_D + (P_V_given_not_D * P_not_D)
print(f"4. Probability of damage detected: {P_V:.4f}")

# 5. Probability that damage exists given the test detected a defect
P_D_given_V = P_V_and_D / P_V
print(f"5. Probability of damage given test detected a defect: {P_D_given_V:.4f}")
```

```python
# prompt: 7.	Multiple Hypothesis Testing
# Let the background information X to be the following statement:
# X ≡ We have 11 automatic machines producing gadgets, which pour out of the machines into 11 boxes. This example corresponds to a very early state in the development of the gadgets, because 10 of the machines produce one in six defective. The 11th machine is even worse; it makes one in three defective. There is also the possibility that something goes entirely wrong in the machine that made our gadgets and it is turning out 99% defective. The output of each machine has been collected in an unlabeled box and stored in the warehouse.
# The prior evidences (hypotheses) for various propositions of interest are as follows:
# A ≡ we choose a bad batch 1/3 defective
# B ≡ we choose a good batch 1/6 defective
# C ≡ we choose a bad batch 99/100 defective
# The prior probability of each hypothesis is defined as:
# The data proposition D would stand for the statement that:
# D ≡ m gadgets were tested and everyone was defective.
# •	Calculate P(D|AX), P(D|BX), P(D|CX)
# •	Calculate P(D|X)
# •	Calculate P(A|DX), P(B|DX), P(C|DX)
# •	Calculate the evidence e(A|DX), e(B|DX), e(C|DX)
# •	Plot in the same graph e(C|DX), e(A|DX) and e(B|DX) as a function of m number of tests (which are all defective).
# •	What is the number of defective results needed in order to accumulate 20 db of information in favor of hypothesis C?

import numpy as np
import matplotlib.pyplot as plt

# Prior probabilities
P_A = 1/11
P_B = 10/11 * (10/11)
P_C = 10/11 * (1/11)

# Likelihoods
def P_D_given_A(m):
    return (1/3)**m

def P_D_given_B(m):
    return (1/6)**m

def P_D_given_C(m):
    return (0.99)**m

# Calculate P(D|X)
def P_D_given_X(m):
    return P_A * P_D_given_A(m) + P_B * P_D_given_B(m) + P_C * P_D_given_C(m)

# Calculate posterior probabilities
def P_A_given_DX(m):
    return P_A * P_D_given_A(m) / P_D_given_X(m)

def P_B_given_DX(m):
    return P_B * P_D_given_B(m) / P_D_given_X(m)

def P_C_given_DX(m):
    return P_C * P_D_given_C(m) / P_D_given_X(m)

# Calculate evidence
def e_A_given_DX(m):
    return np.log10(P_A_given_DX(m) / P_A)

def e_B_given_DX(m):
    return np.log10(P_B_given_DX(m) / P_B)

def e_C_given_DX(m):
    return np.log10(P_C_given_DX(m) / P_C)

# Number of tests
m_values = np.arange(1, 30)

# Calculate evidence for each hypothesis
e_A = [e_A_given_DX(m) for m in m_values]
e_B = [e_B_given_DX(m) for m in m_values]
e_C = [e_C_given_DX(m) for m in m_values]

# Plot the evidence
plt.plot(m_values, e_C, label='e(C|DX)')
plt.plot(m_values, e_A, label='e(A|DX)')
plt.plot(m_values, e_B, label='e(B|DX)')
plt.xlabel('Number of Defective Tests (m)')
plt.ylabel('Evidence (dB)')
plt.title('Evidence for Hypotheses A, B, and C')
plt.legend()
plt.grid(True)
plt.show()

# Find the number of defective results for 20 dB information
target_evidence = 20  # 20 dB
m_20db = 0

for m in m_values:
    if e_C_given_DX(m) >= target_evidence:
        m_20db = m
        break
print(f"Number of defective results needed for 20 dB in favor of hypothesis C: {m_20db}")
```
