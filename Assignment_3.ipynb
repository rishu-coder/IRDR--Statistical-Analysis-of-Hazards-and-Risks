```python
# prompt: First-excursion collapse probability assessment for a structure subjected to
# an aftershock sequence (Geometric Distribution)
# A strong earthquake event is usually followed by a sequence of (often smaller-in-magnitude
# earthquake events called as the Aftershocks). The ensemble of these aftershocks is also referred to
# as a Seismic Sequence.
# Suppose that the probability of structural collapse given that an earthquake of interest has
# happened is equal to   (assuming that each time an earthquake occurs, it hits the intact structure;
# a clearly un-realistic assumption!). Assume that in an interval of time [0,t], n aftershock events
# have taken place.
# • What is the probability that the kth aftershock is going to lead to structural collapse (for
# the first time)?
# A ≡ kth aftershocks leads to structural collapse (kth trial is the first success)
# I ≡ the probability of collapse (success) is 
# • What is the moment generating function for k? (hint:
# the formula for the sum of a geometric series).
# M
#  K
#  t
#  ( )
#  = 
#  k
#  tk
#  e P k
#  ( )
#  , you can use
# • What is the expected value for k? (hint: use the moment generating function).
# • What is the best-estimate for k (e.g., expected value plus one standard deviation)?

import sympy

# Probability of collapse
pi = sympy.Symbol('pi', positive=True)

# Number of aftershocks
k = sympy.Symbol('k', integer=True, positive=True)

# Probability that the kth aftershock leads to the first collapse
probability_kth_collapse = (1 - pi)**(k - 1) * pi

# Moment generating function for k (Geometric Distribution)
t = sympy.Symbol('t')
M_k = sympy.Sum(sympy.exp(t * k) * (1 - pi)**(k - 1) * pi, (k, 1, sympy.oo))

# Simplify the moment generating function
M_k = pi * sympy.exp(t) / (1 - (1 - pi) * sympy.exp(t))


# Expected value of k
E_k = sympy.diff(M_k, t).subs({t: 0})

# Variance of k
M_k_2 = sympy.diff(M_k,t,2).subs({t:0})
Var_k = M_k_2 - E_k**2


# Standard deviation
std_dev_k = sympy.sqrt(Var_k)

# Best estimate of k
best_estimate_k = E_k + std_dev_k

print("Probability that the kth aftershock leads to first collapse:", probability_kth_collapse)
print("Moment Generating Function for k:", M_k)
print("Expected value for k:", E_k)
print("Variance for k:", Var_k)
print("Standard Deviation for k:", std_dev_k)
print("Best estimate for k (E[k] + std_dev[k]):", best_estimate_k)
```

```python
# prompt:  Calculating the expected direct repair costs for the severe damage limit state
# (Negative Binomial Distribution)
# The direct repair costs incurred to a given infrastructure for a given limit state (e.g., Severe
# Damage) are referred to as those costs that are incurred due to direct consequence of a strong
# earthquake for that limit state (i.e., severe damage).
# Suppose that the probability that the infrastructure exceeds the severe damage limit state given that
# an earthquake of interest has happened is equal to   (assuming that each time the structure exceeds
# the severe damage limit state, it is repaired back to its intact state; looking at the entire life of an
# infrastructure, it is somewhat reasonable!). Moreover, assume that the cost of repair associated to
# the severe damage limit state is equal to CD.
# • What is the probability distribution for the number of earthquakes n that need to
# happen in order for the number of repair operations to reach a given value k?
# A ≡ nth events (trial) lead to kth repair (success)
# I ≡ the occurrence of n event (success), and the probability of success of a single trial is 
# • What is the expected value for the number of earthquakes n that need to happen for the
# number of repair operations to reach a given value k? (hint: use the parameters of the
# Geometric distribution and the fact that we have the sum of k independent and
# identically distributed geometric variables)
# • What is the standard deviation for the number of earthquakes n that need to happen for
# the number of repair operations to reach a given value k?
# • Now, imagine that exactly n earthquake events of interest have taken place in time
# interval [0,t]. What is the probability that the infrastructure has to be repaired exactly
# k times during interval [0,t]?
# • What is the expected value for k?
# • What is the expected repair costs fo

import sympy
from sympy.stats import NegativeBinomial, E, variance
import numpy as np

# Parameters
pi = sympy.Symbol('pi', positive=True)  # Probability of exceeding the severe damage limit state
k = sympy.Symbol('k', integer=True, positive=True) # Number of repairs
CD = sympy.Symbol('CD', positive=True) # Cost of repair
n = sympy.Symbol('n', integer=True, positive=True) # Number of earthquakes


# Probability distribution for the number of earthquakes (n) to reach k repairs
# This follows a Negative Binomial distribution
n_dist = NegativeBinomial('N', k, pi)


# Expected value of n
E_n = E(n_dist)

# Standard deviation of n
std_dev_n = variance(n_dist)**0.5


# Probability of exactly k repairs in n earthquake events
# This is given by the PMF of the negative binomial distribution
prob_k_repairs = sympy.binomial(n -1,k-1) * (pi**k) * ((1-pi)**(n-k))

# Expected value of k given n earthquakes (Binomial Distribution)
E_k_given_n = n * pi

# Expected repair costs for the severe damage limit state
E_repair_costs = E_k_given_n * CD

print("Expected value of n (number of earthquakes for k repairs):", E_n)
print("Standard deviation of n:", std_dev_n)
print("Probability of exactly k repairs in n earthquakes:", prob_k_repairs)
print("Expected value for k given n earthquakes:", E_k_given_n)
print("Expected repair costs:", E_repair_costs)

#Example usage with numerical values
pi_val = 0.01
k_val = 5
CD_val = 1000000
n_val = 200

E_n_num = E_n.subs({pi: pi_val, k: k_val})
std_dev_n_num = std_dev_n.subs({pi: pi_val, k: k_val})
prob_k_repairs_num = prob_k_repairs.subs({pi:pi_val, k:k_val, n:n_val})
E_k_given_n_num = E_k_given_n.subs({n:n_val, pi:pi_val})
E_repair_costs_num = E_repair_costs.subs({n: n_val, pi: pi_val, CD:CD_val})

print("Example Calculations")
print("Expected value of n:", E_n_num)
print("Standard deviation of n:", std_dev_n_num)
print("Probability of 5 repairs in 200 events", prob_k_repairs_num)
print("Expected value for k given n = 200 earthquakes:", E_k_given_n_num)
print("Expected repair costs:", E_repair_costs_num)
```
