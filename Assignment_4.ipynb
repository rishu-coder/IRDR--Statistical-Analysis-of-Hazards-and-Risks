The Filtered Poisson
Suppose that the occurrence of earthquakes with Sa(T) > 0.40g is described by a Poisson
distribution with annual rate equal to 0.01. Moreover, suppose that a given structure with
fundamental period equal to T is going to fail with probability 0.1 if it is hit by earthquakes with
Sa(T) > 0.40g.  
• What is the probability of having two earthquakes with Sa(T) > 0.40g in a year?  
• What is the probability of having two earthquakes with Sa(T) > 0.40g in 10 years?
• What is the probability of having at least one earthquake with Sa(T) > 0.40g in 10 years?
• What is the expected number of events in 1000 years?
• What is the best estimate for the number of events in 1000 years?
• Assuming that each time the structure fails, it is immediately repaired (to its original
state), what is the annual rate of structural failure?
• What is the expected number of structural failures in 50 years?
• Assuming a service life of 50 years, what is the probability that the structure does not fail
in its service life?
• What is the probability of failure during the service life of the structure?

```python
#  The Filtered Poisson
# Suppose that the occurrence of earthquakes with Sa(T) > 0.40g is described by a Poisson
# distribution with annual rate equal to 0.01. Moreover, suppose that a given structure with
# fundamental period equal to T is going to fail with probability 0.1 if it is hit by earthquakes with
# Sa(T) > 0.40g.
# • What is the probability of having two earthquakes with Sa(T) > 0.40g in a year?
# • What is the probability of having two earthquakes with Sa(T) > 0.40g in 10 years?
# • What is the probability of having at least one earthquake with Sa(T) > 0.40g in 10 years?
# • What is the expected number of events in 1000 years?
# • What is the best estimate for the number of events in 1000 years?
# • Assuming that each time the structure fails, it is immediately repaired (to its original
# state), what is the annual rate of structural failure?
# • What is the expected number of structural failures in 50 years?
# • Assuming a service life of 50 years, what is the probability that the structure does not fail
# in its service life?
# • What is the probability of failure during the service life of the structure?

import math

# Annual rate of earthquakes with Sa(T) > 0.40g
annual_rate = 0.01

# Probability of structural failure given an earthquake
failure_probability = 0.1

# 1. Probability of two earthquakes in a year
def poisson_probability(k, rate):
  return (math.exp(-rate) * (rate**k)) / math.factorial(k)

prob_two_earthquakes_year = poisson_probability(2, annual_rate)
print(f"Probability of two earthquakes in a year: {prob_two_earthquakes_year}")

# 2. Probability of two earthquakes in 10 years
prob_two_earthquakes_10years = poisson_probability(2, annual_rate * 10)
print(f"Probability of two earthquakes in 10 years: {prob_two_earthquakes_10years}")

# 3. Probability of at least one earthquake in 10 years
prob_at_least_one_10years = 1 - poisson_probability(0, annual_rate * 10)
print(f"Probability of at least one earthquake in 10 years: {prob_at_least_one_10years}")

# 4. Expected number of events in 1000 years
expected_events_1000years = annual_rate * 1000
print(f"Expected number of events in 1000 years: {expected_events_1000years}")

# 5. Best estimate for the number of events in 1000 years (same as expected value for Poisson)
best_estimate_1000years = expected_events_1000years
print(f"Best estimate for the number of events in 1000 years: {best_estimate_1000years}")

# 6. Annual rate of structural failure
annual_failure_rate = annual_rate * failure_probability
print(f"Annual rate of structural failure: {annual_failure_rate}")

# 7. Expected number of structural failures in 50 years
expected_failures_50years = annual_failure_rate * 50
print(f"Expected number of structural failures in 50 years: {expected_failures_50years}")

# 8. Probability of no failure in 50 years
prob_no_failure_50years = math.exp(-annual_failure_rate * 50)
print(f"Probability of no failure in 50 years: {prob_no_failure_50years}")

# 9. Probability of failure during the service life
prob_failure_50years = 1 - prob_no_failure_50years
print(f"Probability of failure during the service life: {prob_failure_50years}")
```

2. Earthquake Hazard Assessment
Suppose that the occurrence of earthquakes with M > 4.5 is described by a Poisson distribution
with annual rate equal to 0.1.
• What is the probability of having two earthquakes with M > 4.5 in a year?  
• What is the probability of having two earthquakes with M > 4.5 in 20 years?
• Assuming that an earthquake with M > 4.5 has just taken place, what is the expected time
to the next earthquake with M > 4.5?
• Assuming that an earthquake with M > 4.5 has just taken place, what is the expected time
to the third earthquake with M > 4.5?
• We know that in the past ten years, there has been no earthquake with M > 4.5, what is
the probability of having at least one earthquake in the next year?
• Imagine that the probability of having PGA > 0.5g given that M > 4.5 is equal to 0.01.
What is the probability of having at least one earthquake with PGA > 0.50g in 10 years?
• What is the expected number of events with PGA > 0.5g in 1000 years?
• Given that 3 earthquakes with M > 4.5 have taken place, what is the probability that all
of them have PGA > 0.5g?

```python
# 2. Earthquake Hazard Assessment
# Suppose that the occurrence of earthquakes with M > 4.5 is described by a Poisson distribution
# with annual rate equal to 0.1.
# • What is the probability of having two earthquakes with M > 4.5 in a year?
# • What is the probability of having two earthquakes with M > 4.5 in 20 years?
# • Assuming that an earthquake with M > 4.5 has just taken place, what is the expected time
# to the next earthquake with M > 4.5?
# • Assuming that an earthquake with M > 4.5 has just taken place, what is the expected time
# to the third earthquake with M > 4.5?
# • We know that in the past ten years, there has been no earthquake with M > 4.5, what is
# the probability of having at least one earthquake in the next year?
# • Imagine that the probability of having PGA > 0.5g given that M > 4.5 is equal to 0.01.
# What is the probability of having at least one earthquake with PGA > 0.50g in 10 years?
# • What is the expected number of events with PGA > 0.5g in 1000 years?
# • Given that 3 earthquakes with M > 4.5 have taken place, what is the probability that all
# of them have PGA > 0.5g?

import math

# Annual rate of earthquakes with M > 4.5
annual_rate_m45 = 0.1

# Probability of PGA > 0.5g given M > 4.5
prob_pga_given_m45 = 0.01

# 1. Probability of two earthquakes with M > 4.5 in a year
prob_two_m45_year = poisson_probability(2, annual_rate_m45)
print(f"Probability of two earthquakes (M > 4.5) in a year: {prob_two_m45_year}")

# 2. Probability of two earthquakes with M > 4.5 in 20 years
prob_two_m45_20years = poisson_probability(2, annual_rate_m45 * 20)
print(f"Probability of two earthquakes (M > 4.5) in 20 years: {prob_two_m45_20years}")

# 3. Expected time to the next earthquake with M > 4.5
expected_time_next_m45 = 1 / annual_rate_m45
print(f"Expected time to the next earthquake (M > 4.5): {expected_time_next_m45} years")

# 4. Expected time to the third earthquake with M > 4.5
expected_time_third_m45 = 3 / annual_rate_m45
print(f"Expected time to the third earthquake (M > 4.5): {expected_time_third_m45} years")

# 5. Probability of at least one earthquake in the next year (given no earthquakes in 10 years)
# This uses the memoryless property of the Poisson process
prob_at_least_one_next_year = 1 - math.exp(-annual_rate_m45)
print(f"Probability of at least one earthquake (M > 4.5) in the next year: {prob_at_least_one_next_year}")

# 6. Probability of at least one earthquake with PGA > 0.5g in 10 years
annual_rate_pga = annual_rate_m45 * prob_pga_given_m45
prob_at_least_one_pga_10years = 1 - math.exp(-annual_rate_pga * 10)
print(f"Probability of at least one earthquake with PGA > 0.5g in 10 years: {prob_at_least_one_pga_10years}")

# 7. Expected number of events with PGA > 0.5g in 1000 years
expected_pga_1000years = annual_rate_pga * 1000
print(f"Expected number of events with PGA > 0.5g in 1000 years: {expected_pga_1000years}")

# 8. Probability that all 3 earthquakes have PGA > 0.5g
prob_all_3_pga = prob_pga_given_m45 ** 3
print(f"Probability that all 3 earthquakes have PGA > 0.5g: {prob_all_3_pga}")
```

4. Seismic Risk Assessment
Suppose that the occurrence of earthquakes with M > 5.0 is described by a Poisson distribution
with annual rate equal to 0.2.
• What is the probability of having at least one earthquake with M > 5.0 in one year?
• We know that in the last five years, no earthquakes with M > 5.0 have happened. What
is the probability that no earthquakes with M > 5.0 will happen in the next year?
• Consider that the probability of collapse for a structure subjected to an earthquake with
M > 5.0 is equal to 5%. What is the mean annual rate of collapse?
• What is the probability of having at least a collapse in the life cycle of the structure
(tlife=30 years)?
• We know that the structure was built 25 years ago, and it has never collapsed. What is
the probability that the structure will collapse in the next five years?
• We know that the structure was built 25 years ago. 20 years ago, it was destroyed and
immediately reconstructed. How many times it has to wait on average before observing
the first collapse?

```python
# 4. Seismic Risk Assessment
# Suppose that the occurrence of earthquakes with M > 5.0 is described by a Poisson distribution
# with annual rate equal to 0.2.
# • What is the probability of having at least one earthquake with M > 5.0 in one year?
# • We know that in the last five years, no earthquakes with M > 5.0 have happened. What
# is the probability that no earthquakes with M > 5.0 will happen in the next year?
# • Consider that the probability of collapse for a structure subjected to an earthquake with
# M > 5.0 is equal to 5%. What is the mean annual rate of collapse?
# • What is the probability of having at least a collapse in the life cycle of the structure
# (tlife=30 years)?
# • We know that the structure was built 25 years ago, and it has never collapsed. What is
# the probability that the structure will collapse in the next five years?
# • We know that the structure was built 25 years ago. 20 years ago, it was destroyed and
# immediately reconstructed. How many times it has to wait on average before observing
# the first collapse?

# Annual rate of earthquakes with M > 5.0
annual_rate_m5 = 0.2

# Probability of collapse given an earthquake with M > 5.0
collapse_probability = 0.05

# 1. Probability of at least one earthquake with M > 5.0 in one year
prob_at_least_one_m5 = 1 - poisson_probability(0, annual_rate_m5, 1)
print(f"Probability of at least one earthquake (M > 5.0) in one year: {prob_at_least_one_m5}")

# 2. Probability of no earthquakes in the next year (given no earthquakes in the last 5 years)
# Poisson process is memoryless
prob_no_m5_next_year = math.exp(-annual_rate_m5 * 1)
print(f"Probability of no earthquakes (M > 5.0) in the next year: {prob_no_m5_next_year}")

# 3. Mean annual rate of collapse
annual_collapse_rate = annual_rate_m5 * collapse_probability
print(f"Mean annual rate of collapse: {annual_collapse_rate}")

# 4. Probability of at least one collapse in 30 years
prob_at_least_one_collapse_30years = 1 - math.exp(-annual_collapse_rate * 30)
print(f"Probability of at least one collapse in 30 years: {prob_at_least_one_collapse_30years}")

# 5. Probability of collapse in the next 5 years (given no collapse in the last 25 years)
# Memoryless property applies
prob_collapse_next_5years = 1 - math.exp(-annual_collapse_rate * 5)
print(f"Probability of collapse in the next 5 years: {prob_collapse_next_5years}")

# 6. Expected time to the first collapse after reconstruction (20 years ago)
# Memoryless property applies - effectively a new structure
expected_time_to_first_collapse = 1 / annual_collapse_rate
print(f"Expected time to the first collapse after reconstruction: {expected_time_to_first_collapse} years")
```
